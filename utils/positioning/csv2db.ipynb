# Сбор статистики в виде БД

Здесь рассматриваются вопросы сбора статистики по силе сигнала сотовой связи в виде БД, удобной для дальнейшего использования в системе позиционирования.

## Описание БД

### Описание структуры имеющихся данных

#### Общие сведения
Метрополитен с точки зрения пассажира представляет собой линейно-протяженный объект, состоящий из нескольких пересекающихся линий. Каждая линия состоит из нескольких сегментов (участков туннелей или переходов), которые образуются станциями метро. Примером такого сегмента может быть участок туннеля от станции "Академическая" до станции "Ленинский проспект". Движение в метро происходит в разных направлениях, поэтому участок тунелля от станции "Ленинский проспект" до "Академической" -- не тот же самый сегмент, а другой.

При этом:
 - Все линии метро поделены на сегменты. Точки разреза пронумерованы уникальными идентификаторами.
 - Все сегменты пронумерованы двумя числами -- (номер начального идентификатора сегмента, номер конечного идентификатора сегмента).

Поэтому для позиционирования пассажира в метро достаточно указать номер сегмента и точку на этом сегменте. Точку можно задать одним числом от нуля до единицы, означающим долю сегмента, в котором эта точка его делит. Например, ноль означает, что точка позиционирования находится в начале сегмента, один -- в конце сегмента, 0.5 -- ровно в середине и т.п.

Альтернативный подход (который здесь не использован) подразумевает использование реальных расстояний от начала сегмента в метрах. Он более сложен в реализации, чем подход через отношение длин отрезков, поскольку требует знания реальных длин и координат. В тоже время подход через отношение позволяет работать как в условной системе координат, так и с реальными физическими координатами: зная длину сегмента от оношения легко перейти к реальным расстояниям.


#### Позиционирование с использованием сигналов сотовой сети
В метро не работают системы глобального позиционирования и очевидной альтернативой является вычисление местоположения по силе сигналов сотовой сети.

Таким образом перед нами стоит задача: создать функцию, которая по имеющимся данным о силе сигнала сотовой сети того или иного оператора возвращала позиционирующую информацию (номер сегмента и точку на этом сегменте). Помимо силы сигнала опционально имеется ряд иных вспомогательных величин: показания акселерометра, гироскопа и т.п.

#### Имеющиеся данные
На данный момент собранные данные о силе сигнала и показания различных сенсоров хранятся в репозитории https://github.com/nextgis/metrocell пример данных можно посмотреть тут: https://github.com/nextgis/metrocell/tree/master/data/proc/msk/cell и https://github.com/nextgis/metrocell/tree/master/data/proc/msk/sensor.

Внутри каждого файла находится список строк, в которых указано время замера от начала движения и показатели измеряемой величины.

### Структура БД

Таким образом имеем, что для использования силы сигнала сотовой сети как предиктора, нужно предварительно построить обучающую выборку из имеющихся данных в формате: (местоположение; показатели сенсоров).

Местоположение задается номером сегмента (есть в данных) и координатой внутри сегмента числом от нуля до единицы. Коодината точки в сегменте должна быть вычислена тем или иным образом (об этом позднее, в соотвествующем разделе).

Итак, обучающая выборка будет создаваться в виде таблицы БД следущего вида:

|№ записи | № сегмента | координата в сегменте | показания сенсора1 |показания сенсора2 | ... | показания сенсораN|
|---------|------------|-----------------------|--------------------|-------------------|-----|-------------------|
|...|...|...|...|...|...|...|


### Получение координаты внутри сегмента

Итак, в качестве исходных данных имеем таблицу показаний сенсоров, сделанных при проходе по сегменту. Среди показаний есть время, затраченное с момента начала движения, показания акселеромметров и т.д. Эти данные позволяют оценить значение координаты в момент регистрации показаний сенсоров, поэтому можно написать функцию, которая будет назначать записям из таблицы координаты внутри сегмента.

В то же время реализация конкретной функции -- вопрос отдельный, который будет рассматриваться в соответствующем разделе. Для целей описания структуры БД и первичного сбора статистики достаточно определить общий вид функции: ее назначение, аргументы и формат выходных данных.

#### Интерфейс функции `add_ratio()`

На вход функция будет принимать таблицу данных, и словарь, который содержит перечисление колонок, существенных для определения положения точки замера в сегменте. На выходе эта функция выдает ту же таблицу с данными, но в ней появляется дополнительные колонки: 
  * `ratio`, которя содержит число из диапазона [0, 1] -- отношение в котором точка замера делит сегмент;
  * `seg_begin` и `seg_end` -- колонки, хранящие идентификаторы точек начала и конца сегмента.

#### Реализация

В качестве заглушки будем использовать простую реализацию функциии `add_ratio()`, которая расчитывает координаты точки исходя лишь из времени проведенного в движении. При этом движение считается равномерным (понятно, что это не так, но более точная реализация функции будет сделана позднее) -- в первом приближении это позволит расставить показания замеров на сегменте. После того, как будет написана другая реализация функции, потребуется произвести перерасчет процедуры заполняющей БД данными.

## Программная реализация

В этом разделе собран програмный код, реализующий описанные выше функции.

```python
def _prepare_frame(url):
    '''Вспомогательная функция:
    считывает данные в виде датафрейма, чистит их от мусора,
    преобразовывает типы
    '''
    frame = pd.io.parsers.read_csv(url)
    
    # Приведем время к читаемому виду
    frame['TimeStamp'] = frame['TimeStamp'] / 1000  # время там в милисекундах
    ts = frame['TimeStamp'] 
    ts = ts.apply(datetime.datetime.fromtimestamp)
    frame['Time'] = ts

    # Удалим колонки, которые не нужны
    frame.drop(['User', 'Name', 'TimeStamp', 'Active'], axis=1, inplace=True)
    
    return frame
```

```python
def add_ratio(frame, description={'Time': 'Time', 'ID': 'ID'}):
    """Вставить в фрейм данные о начале и конце сегментов и координаты замера в сегменте
    
    description -- словарь, в котором определяются названия колонок, хранящих:
        Time: колонка с датой и временем замера
        ID:   колонка с идентификаторами точек начал и концов сегментов
    """
    # Чтобы не испортить данные
    frame = frame.copy()
    
    time_col = description['Time']
    id_col = description['ID']
    
    # Cортируем по времени для надежности:
    frame = frame.sort(columns=time_col)

    size = len(frame[id_col])

    # Мы опираемся на то, что в таблице записи 
    # сделаны последовательно -- в первой записи стоит
    # номер точки начала движения, затем идет вторая запись и т.д.
    seg_begin = pd.Series(np.zeros(size), dtype=np.int)
    seg_end = pd.Series(np.zeros(size), dtype=np.int)
    id_beg = frame[id_col].iloc[0]
    id_end = frame[id_col].iloc[-1]
    for i in range(len(seg_begin)):
        if not np.isnan(frame[id_col].iloc[i]):
            id_beg = frame[id_col].iloc[i]
        if not np.isnan(frame[id_col].iloc[size - i - 1]):
            id_end = frame[id_col].iloc[size - i - 1]

        seg_begin.iloc[i] = id_beg
        seg_end.iloc[size - i - 1] = id_end

    frame['seg_begin'] = seg_begin
    frame['seg_end'] = seg_end

    time_ser = frame[time_col]
    start, stop = time_ser.iloc[0], time_ser.iloc[-1]

    # Движение примем равномерным
    ratio = (time_ser - start)/(stop-start)
    frame['ratio'] = ratio
    
    # Удалим колонки, которые больше не нужны
    frame.drop([id_col, time_col], axis=1, inplace=True)
    
    return frame
```

```python
import pandas as pd
import numpy as np

from count_stats_by_path import (
  get_file_list, 
  get_begin_id,
  get_end_id,
  path_is_stop,
  describe_file,
  get_stat,
  get_geojson
)

OWNER = 'nextgis'
REPO = 'metrocell'
PATH = 'data/proc/msk/cell'
BASE_URL = 'https://api.github.com/repos'

files_description = get_file_list(BASE_URL, OWNER, REPO, PATH)

database = pd.DataFrame()
for url_descr in files_description:
    url = url_descr['download_url']
    print '.',
    frame = _prepare_frame(url)
    frame = add_ratio(frame)
    database = pd.concat([database, frame])

database.to_csv('database.csv')
print 'Saved!'
```
