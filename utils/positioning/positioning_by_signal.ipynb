# Позиционирование по силе сигнала и идентификатору базовой станции

## Исходные данные

Есть БД в формате SQLite, которая содержит данные логов. Они созданы на базе [логов](https://github.com/nextgis/metrocell/tree/master/data/proc/msk/cell), с небольшой предобработкой:

 * Удалены лишние колонки.
 * Добавлены координаты точки замера сигнала.
 * Интервалы между замерами сигнала на каждом сементе сделаны равными одной секунде. Сила сигнала и координаты проинтерполированы, чтобы соотвествовать точкам замера.
 
Колонки БД:
 
 * CID, LAC, MNC, MCC -- идентификатор станции.
 * X,Y -- координаты точки замера.
 * Power -- сила сингала.
 * Time, Date -- дата и время замера силы сигнала
 * SEG -- символьная метка сегмента.
 * obs_id -- номер лога, из которого были импортированы данные (строки с разными obs_id были собраны в разные поездки).

 Пример содержимого БД:

## Моделирование обработки

Промоделируем ситуацию, когда человек едет в вагоне и собирает информацию о базовой станции и силе сигнала. Ему нужно определиться с тем, где он сейчас находится.

### Выбор логов для тестового и обучающего множеств

Чтобы это смоделировать извлечем из БД несколько логов -- они будут использоваться как новые сигналы, которые приходят с телефона пользователя (т.е. считаем, что они были ранее неизвестны системе позиционирования).

Определимся, какие логи мы можем использовать в качестве таких контрольных логов. Выберем из БД все уникальные базовые станции и подсчитаем количество поездок по ним. Отобразим информацию по тем из сегментов, по которым было собрано 4 и более логов:

```python
sql = """
SELECT count(distinct(obs_id)) as travels, lac, cid, mnc, mcc, seg
FROM observation
GROUP BY lac, cid, mnc, mcc, seg
HAVING travels > 3 AND cid > 0
ORDER BY travels DESC
"""
frame = pd.read_sql_query(sql, engine)
frame
```

Из таблицы видно, что самые частые логи приходятся на сегмент 099-100 для базовой станции (Lac/Cid) 770/543 оператора МТС. Кроме того, эта базовая станция функционирует еще на нескольких сегментах: 100-101, 098-099, 099-098, 100-099, 101-100. Поэтому в качестве примера будем использовать именно эту БС:

 * она покрывает значительное пространство и позиционирование только по LAC/CID мало что дает;
 * для нее собрана большая статистика.

### Прогноз позиции по силе сигнала

Предположим, что пользователь "прицепился" к БС LAC=770, CID=543, MNC=1. Покажем на схеме места, где он может ехать:

```python
sql = 'SELECT * FROM observation WHERE lac=770 AND cid=543 AND mnc=1'

frame = pd.read_sql_query(sql, engine)

figure(num=None, figsize=(10, 6))
plt.plot(frame.x, frame.y, '.')
plt.title("Observations for LAC=770, CID=543, MNC=1")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

Посмотрим номера логов, в которых зафиксирован LAC=770, CID=543:

```python
sql = 'SELECT distinct(obs_id) FROM observation WHERE lac=770 AND cid=543 AND mnc=1'

obs_id = pd.read_sql_query(sql, engine)
obs_id
```

Итого получили 40 логов по БС LAC=770, CID=543.

Извлечем один из них в тестовое множество, а остальные положим в обучающее.

```python
test_id = np.random.choice(obs_id.obs_id)

sql = "SELECT * FROM observation WHERE obs_id = %s ORDER BY time" % (test_id, )
test = pd.read_sql_query(sql, engine)
test.head()

sql = "SELECT * FROM observation WHERE obs_id <> %s ORDER BY obs_id, time" % (test_id, )
train = pd.read_sql_query(sql, engine)
```

Предположим, что пользователь включил телефон в случайный момент времени и получил данные по силе сигнала за N секунд.

```python
N = 10

size = test.shape[0]

id0 = np.random.choice(size) # Момент включения
id1 = min(id0 + N, size)

t1 = test.loc[id0:id1,:]
t1
```

Определим зарегистрированные телефоном идентификаторы БС и отобразим на карте те места, в которых эти БС используются:

```python
bs = (t1[['CID', 'LAC', 'MNC', 'MCC']]).drop_duplicates()

suspect = None
for row in bs.iterrows():
    cid, lac, mnc, mcc = row[1]
    sql = 'SELECT * FROM observation WHERE lac=%s AND cid=%s AND mnc=%s AND mcc=%s' %(lac, cid, mnc, mcc)
    tmp = pd.read_sql_query(sql, engine)
    if suspect is None:
        suspect = tmp
    else:
        # найдем участки, где встречаются все зафиксированные id
        # (пересечение множеств сегментов)
        known_seg = set(pd.unique(suspect.SEG))
        new_seg = set(pd.unique(tmp.SEG))
        common = known_seg.intersection(new_seg)
        suspect = pd.concat([suspect, tmp])
        suspect = suspect[suspect.SEG.isin(common)]

print 'Segments:', pd.unique(suspect.SEG)  

figure(num=None, figsize=(10, 6))
plt.plot(suspect.x, suspect.y, '.')
plt.title("Observations for LAC=%s, CID=%s, MNC=%s" % (lac, cid, mnc))
plt.xlabel("X")
plt.ylabel("Y")
plt.show()   
```
